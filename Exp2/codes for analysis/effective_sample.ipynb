```python
def eff_n_less_500(list):
    n = 0
    for i in range(len(list)):
        if list[i]<=500:
            n+=1
    return n

def r_hat_more_101(list):
    n = 0
    for i in range(len(list)):
        if list[i]>=1.01:
            n+=1
    return n
def mean_n_eff(list):
    list_mean_n_eff = []
    for i in range(len(list)):
        list_mean_n_eff.append(round(np.mean(list[i])))
    return list_mean_n_eff
def mean_r_hat(list):
    list_mean_r = []
    for i in range(len(list)):
        mean_r = 0
        for j in range(len(list[i])):
            mean_r+=(1+np.abs(list[i][j]-1))
        list_mean_r.append(round(mean_r/len(list[i]),4))
    return list_mean_r
```
